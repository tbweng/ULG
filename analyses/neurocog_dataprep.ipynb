# psytoolkit
Note: redcap import performed in separate scripts

```python
# first clean data
files = glob('./psytoolkit/*/data.csv')

for f in files: 
    
    # read in csv to count number of values in header
    df = pd.read_csv(f, header = None, nrows=1) 
    ncols = len(df.columns)
    
    # re-read limit ncols
    df = pd.read_csv(f, usecols = range(ncols))
    df['TIME_start'] = pd.to_datetime(df['TIME_start'])
    
    # exclude data before 8/20/19
    df = df[(df['TIME_start'] >= dt.datetime(2019, 8, 20))]

    # drop rows if any column is blank
    df.dropna(how = 'any', inplace=True)

    # make all pt ids consisitent
    df['participant_info:1'] = df['participant_info:1'].str.replace('ulg_0','')

    # zero pad pt id and add 'ulg_' string
    df['participant_info:1'] = 'ulg_' + df['participant_info:1'].astype(str).apply(lambda x: x.zfill(3))
    
    # write out
    df.to_csv(f, index = False)
```
